## Cleaning and Preprocessing Data
- Here we clean the `df['Duration]` column by converting it to numerical form
- There were some entries where the data was in ranges i.e., 2-5.1 years. For that I have simply taken the mean of those entries

````python
duration_years:list[float] = []
ctr : int = 0
for x in df['Duration']:
    if x[3] != '-' and x[1] != '-' and x[2] != '-' and x[4] != '-':
        duration_years.append(float(x[-6::-1]))
    else:
        p,q = map(float, x[-6::-1].split("-"))
        duration_years.append((p+q)/2)
        ctr += 1
print(f"{ctr} records were modified.")
````

````python
# mean

def df_mean(dataframe:pd.DataFrame) -> pd.DataFrame:
    mean_dict = {}
    for col in dataframe.columns:
        if pd.api.types.is_numeric_dtype(dataframe[col]):
            cum_sum : float = 0
            count : int = 0
            for x in df[col].values:
                cum_sum += x
                count += 1
            mean_dict[col] = [cum_sum/count]
        # else:
        #     print(f'Skipping {col} as it isn\'t numeric.')
    mean_list = pd.DataFrame(mean_dict, index=['mean'])
    return mean_list
    
````

````python
# Quick sort in Python - 
# Ackchually  using this code is a very bad idea as it is in python and the interpreter will have to incurr additional overhead 
# so before a homie tells me sht about my unecessary yapping I'm just letting you know in advance

# TL;DR - Use sort() it is much faster as it is implemented in C 
def partition(array, low, high):
  pivot = array[high]
  i = low - 1
  for j in range(low, high):
    if array[j] <= pivot:
      i = i + 1
      (array[i], array[j]) = (array[j], array[i])
  (array[i + 1], array[high]) = (array[high], array[i + 1])
  return i + 1

def quickSort(array, low, high):
  if low < high:
    pi = partition(array, low, high)
    quickSort(array, low, pi - 1)
    quickSort(array, pi + 1, high)
````

````python
def df_median(dataframe : pd.DataFrame) -> pd.DataFrame:
    sorted_list = []
    median_dict = {}
    for col in dataframe.columns:
        if pd.api.types.is_numeric_dtype(dataframe[col]):
            sorted_list = dataframe[col].values
            quickSort(sorted_list, 0, len(sorted_list) - 1)  # sorts it inplace
            if len(sorted_list) % 2:
                mid = (len(sorted_list) + 1) // 2
                median_dict[col] = sorted_list[mid]
            else:
                mid_1 = len(sorted_list) // 2
                mid_2 = (len(sorted_list) - 1) // 2
                median_dict[col] = (sorted_list[mid_1] + sorted_list[mid_2]) / 2
        else:
            print(f'Skipping {col} as it isn\'t numeric')
    median_list = pd.DataFrame(median_dict, index=['median'])
    return median_list
````

````python
def df_min(dataframe:pd.DataFrame)-> pd.DataFrame:
    min_dict = {}
    for col in dataframe.columns:
        if pd.api.types.is_numeric_dtype(dataframe[col]):
            local_min = float('inf')
            for x in dataframe[col].values:
                if x < local_min:
                    local_min = x
            min_dict[col] = local_min
        else:
            print(f'Skipping {col} as it isn\'t numeric')
    return pd.DataFrame(min_dict, index=['min'])
````

````python
def df_max(dataframe:pd.DataFrame)-> pd.DataFrame:
    max_dict = {}
    for col in dataframe.columns:
        if pd.api.types.is_numeric_dtype(dataframe[col]):
            local_max = float('-inf')
            for x in dataframe[col].values:
                if x > local_max:
                    local_max = x
            max_dict[col] = local_max
        else:
            print(f'Skipping {col} as it isn\'t numeric')
    return pd.DataFrame(max_dict, index=['min'])
````

````python
def df_std(dataframe:pd.DataFrame)-> pd.DataFrame:
    std_dict = {}
    for col in dataframe.columns:
        if pd.api.types.is_numeric_dtype(dataframe[col]):
            sum_values = 0.0  # Sum of all values
            count = 0         # Number of values
            for x in dataframe[col].values:
                sum_values += x
                count += 1
            mean = sum_values / count
            sum_squared_diff = 0.0
            for x in dataframe[col].values:
                sum_squared_diff += (x - mean) ** 2
            std_dev = (sum_squared_diff / count) ** 0.5
            std_dict[col] = std_dev
        else:
            print(f'Skipping {col} as it isn\'t numeric')
    return pd.DataFrame(std_dict, index=['std'])
````
