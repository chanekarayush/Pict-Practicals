## Q6. Turn categorical variables into quantitative variables in Python

### Categorical Values
- intensity
- smoking_status
- activity_type (we can code these into 10 clear categories ranging from 0-9)
- gender

```python
df['gender'] = df['gender'].replace({'M':1, "F":0})
```

```python
df.sample(5)
```

```python
df['intensity'] = df['intensity'].replace({'Low':0,'Medium':1,'High':2})
```

```python
df.sample(5)
```

```python
df['smoking_status'] = df['smoking_status'].replace({'Never':0, 'Former':1, 'Current':2})
```

```python
df.sample(5)
```

```python
df['activity_type'] = df['activity_type'].replace({'Dancing':0, 'Swimming':1, 'Weight Training':2, 'HIIT':3, 'Running':4,
       'Walking':5, 'Tennis':6, 'Basketball':7, 'Yoga':8, 'Cycling':9})
```

```python
df.sample(5)
```
