**Compute Estimators of the main statistical measures like Mean, Variance, Standard Deviation, Covariance, Correlation and Standard error with respect to any example.** \
Display graphically the distribution of samples.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import statistics
```

```python
df = sns.load_dataset("mpg")
df.head()
```

```python
df.mean(numeric_only=True)
```

```python
df.filter(items=[col for col in df.columns if col != 'weight']).mean(numeric_only=True)
```

```python
df.filter(items=[col for col in df.columns if col != 'weight']).mean(numeric_only=True).plot(kind='line', figsize=(10,6), xlabel='Car Attributes', ylabel='Values', title='Car Statistics')
```

```python
print(df.median(numeric_only=True))
print()
print(df.mean(numeric_only=True))
```

```python
statistics.variance(df["mpg"])
```

```python
# sns.violinplot(x=df['mpg'],)
sns.boxplot(x=df["mpg"])
```

```python
statistics.pstdev(df["mpg"])
```

```python
stdev= np.std(df["displacement"])
sns.boxplot(df["displacement"])
```

```python
stdev
```

```python
# covariance
x = [2,4,6,8,10]
y = [3,7,10,14,17]
```

```python
statistics.covariance(x,y)
```

```python
statistics.covariance(df['mpg'], df['displacement'])
```

```python
sns.scatterplot(x=df['mpg'], y=df['displacement'])
```

```python
# correlation
np.corrcoef(x,y)
```

```python
sns.heatmap(np.corrcoef(x,y),annot=True, fmt=".2f", linewidths=0.5)
```

```python
# standard error
statistics.stdev(x) / (len(x)**0.5)
```

```python
import matplotlib.pyplot as plt
```

```python
x_err = statistics.stdev(x) / (len(x)**0.5)
y_err = statistics.stdev(y) / (len(y)**0.5)
```

```python
plt.errorbar(x,y, xerr=x_err, yerr=y_err, fmt='o')
plt.plot(x,y)
plt.show()
```
