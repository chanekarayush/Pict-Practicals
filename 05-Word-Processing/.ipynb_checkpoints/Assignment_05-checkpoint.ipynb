```python
import re
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

from collections import Counter
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

### 1. Load data 

```python
CSV_PATH = "./twitter_data.csv"
TEXT_COL = "Tweet"

df = pd.read_csv(CSV_PATH)
assert TEXT_COL in df.columns, f"Column '{TEXT_COL}' not found. Available: {list(df.columns)}"
print("Rows:", len(df))
print(df[[TEXT_COL]].head())
```

### 2. Pre-processing

```python
stop_words = set(stopwords.words("english"))
lemm = WordNetLemmatizer()

def clean_tweet(s):
    s = str(s)
    s = s.lower()
    s = re.sub(r"http\S+", " ", s)     # URLs
    s = re.sub(r"@\w+", " ", s)        # mentions
    s = re.sub(r"#\w+", " ", s)        # hashtags (remove the token; keep word if you prefer)
    s = re.sub(r"&amp;", " and ", s)   # HTML entity
    s = re.sub(r"[^a-z\s]", " ", s)    # keep only letters and spaces
    s = re.sub(r"\s+", " ", s).strip()
    return s

def tokenize_lemmatize(s):
    toks = s.split()
    toks = [t for t in toks if t not in stop_words and len(t) > 2]  # drop stopwords/short tokens
    toks = [lemm.lemmatize(t) for t in toks]
    return toks

df["clean"] = df[TEXT_COL].apply(clean_tweet)
df["tokens"] = df["clean"].apply(tokenize_lemmatize)
```

```python
print("Example tokens:", df["tokens"].head(3).tolist())
```

### 3. Basic EDA

#### 3.1 Corpus-level Stats

```python
all_tokens = [t for toks in df["tokens"] for t in toks]
num_docs = len(df)
num_tokens = len(all_tokens)
vocab = set(all_tokens)
vocab_size = len(vocab)

print(f"Documents: {num_docs}")
print(f"Total tokens: {num_tokens}")
print(f"Vocabulary size: {vocab_size}")
```

```python
freqs = Counter(all_tokens)
print("Top 25 words:", freqs.most_common(25))
```

### 4. Zipf's Law (rankâ€“frequency)

#### 4.1. rank words by frequency

```python
sorted_items = freqs.most_common()
ranks = np.arange(1, len(sorted_items) + 1)
freq_values = np.array([f for _, f in sorted_items], dtype=float)
```

#### 4.2. log-log

```python
log_ranks = np.log(ranks)
log_freqs = np.log(freq_values)
```

#### 4.3. Linear fit on log-log to estimate slope (~ -1 for Zipf)

```python
slope, intercept = np.polyfit(log_ranks, log_freqs, 1)
print(f"Zipf log-log fit: slope={slope:.3f}, intercept={intercept:.3f}")
```

#### 4.4. Plot

```python
plt.figure()
plt.scatter(log_ranks, log_freqs, s=8)
plt.plot(log_ranks, slope*log_ranks + intercept)
plt.xlabel("log(rank)")
plt.ylabel("log(frequency)")
plt.title("Zipf's Law (rankâ€“frequency, logâ€“log)")
plt.show()
```

### 5. Heaps' Law (vocabulary growth)
> V(N) ~ k * N^Î²
#### Sweep through corpus and record vocab size vs tokens


```python
seen = set()
v_sizes = []
n_tokens_progress = []
count = 0
for toks in df["tokens"]:
    for t in toks:
        count += 1
        seen.add(t)
        if count % 500 == 0:  # sample every 500 tokens to speed up
            n_tokens_progress.append(count)
            v_sizes.append(len(seen))
```

> #### Fit log(V) = log(k) + Î² * log(N)

```python
logN = np.log(np.array(n_tokens_progress, dtype=float))
logV = np.log(np.array(v_sizes, dtype=float))
beta, logk = np.polyfit(logN, logV, 1)
k = math.exp(logk)
print(f"Heaps parameters: k={k:.3f}, beta={beta:.3f}")
```

#### Plot Heaps (linear)

```python
plt.figure()
plt.plot(n_tokens_progress, v_sizes)
plt.xlabel("Tokens (N)")
plt.ylabel("Vocabulary size (V)")
plt.title("Heaps' Law (vocabulary growth)")
plt.show()
```

#### Plot Heaps (logâ€“log)

```python
plt.figure()
plt.scatter(logN, logV, s=10)
plt.plot(logN, beta*logN + logk)
plt.xlabel("log(N)")
plt.ylabel("log(V)")
plt.title("Heaps' Law (logâ€“log)")
plt.show()
```

### 6. Topic Identification (LDA) 
#### Weâ€™ll join tokens to strings to control preprocessing we already did.


```python
docs_clean = df["tokens"].apply(lambda toks: " ".join(toks)).tolist()

vectorizer = CountVectorizer(
    max_features=10000,
    min_df=5,            # ignore very rare terms
    max_df=0.5           # ignore overly common terms
)
```

```python
X = vectorizer.fit_transform(docs_clean)
```

#### 6.1. Fit LDA

```python
N_TOPICS = 10      # adjust as needed
N_TOP_WORDS = 12   # words per topic to display

lda = LatentDirichletAllocation(
    n_components=N_TOPICS,
    random_state=42,
    learning_method="batch",
)
lda.fit(X)
```

```python
terms = np.array(vectorizer.get_feature_names_out())
```

```python
def print_topics(model, terms, n_top_words=12):
    for i, comp in enumerate(model.components_):
        top_idx = np.argsort(comp)[::-1][:n_top_words]
        print(f"\nTopic {i:02d}:")
        print(", ".join(terms[top_idx]))
```

```python
print_topics(lda, terms, n_top_words=N_TOP_WORDS)
```

### 7. Topic distribution for a few documents

```python
topic_dist = lda.transform(X)
print("\nDoc â†’ top topic (first 5):")
for i, row in enumerate(topic_dist[:5]):
    print(i, np.argmax(row), f"(strength={row.max():.3f})")
```
